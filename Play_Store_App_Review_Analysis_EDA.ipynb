#  📌 **Project Summary: Google Play Store Apps Analysis**

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Remove duplicates
apps.drop_duplicates(inplace=True)

# Clean Installs
apps['Installs'] = apps['Installs'].astype(str).str.replace('[+,]', '', regex=True)
apps['Installs'] = pd.to_numeric(apps['Installs'], errors='coerce')

# Clean Price
apps['Price'] = apps['Price'].astype(str).str.replace('$', '', regex=True)
apps['Price'] = pd.to_numeric(apps['Price'], errors='coerce')

# Convert Reviews
apps['Reviews'] = pd.to_numeric(apps['Reviews'], errors='coerce')

# Handle missing values
apps['Rating'].fillna(apps['Rating'].median(), inplace=True)

# Drop critical null rows
apps.dropna(subset=['Installs','Reviews'], inplace=True)

# Log transform
apps['Installs_log'] = np.log1p(apps['Installs'])
apps['Reviews_log'] = np.log1p(apps['Reviews'])

# Revenue estimate
apps['Revenue_Estimate'] = apps['Price'] * apps['Installs']

print("Data Wrangling Completed ✅")
````

**Manipulations Done**

*   Removed duplicates
*   Cleaned numeric columns
*   Applied median imputation
*   Log transformed skewed variables
*   Created revenue estimate metric
